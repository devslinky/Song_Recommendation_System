```python
client_id =  '01b87eb4c04640c89b463737e159089a'
client_secret = 'd6d7f6663f1e4d00ad4611960f283b0c'
import requests
import base64

def get_access_token(client_id, client_secret):
    auth_url = 'https://accounts.spotify.com/api/token'
    auth_header = {
        'Authorization': f'Basic {base64.b64encode((client_id + ":" + client_secret).encode()).decode()}'
    }
    auth_data = {
        'grant_type': 'client_credentials',
        'client_id': client_id,
        'client_secret': client_secret
    }

    auth_response = requests.post(auth_url, data=auth_data, headers=auth_header)
    auth_response_data = auth_response.json()
    access_token = auth_response_data['access_token']
    return access_token

access_token = get_access_token(client_id, client_secret)
print(access_token)
```

```python
def get_release_date(track_id, access_token):

    track_url = f'https://api.spotify.com/v1/tracks/{track_id}'

    headers = {
        'Authorization': f'Bearer {access_token}'
    }

    response = requests.get(track_url, headers=headers)

    if response.status_code == 200:
        track_info = response.json()
        release_date = track_info['album']['release_date']
        return release_date
    else:
        print(f"Error retrieving track details for ID {track_id}: {response.status_code}")
        return None

track_id = '11dFghVXANMlKmJXsNCbNl'
release_date = get_release_date(track_id, access_token)
print("Release Date:", release_date)
```

```python
import time
import pandas as pd

# Check if a saved DataFrame exists (if interrupted previously)
try:
    df = pd.read_csv('output.csv')
    start_idx = df.dropna(subset=['release_date']).shape[0]  # Find the last row with a release date
    print(f"Resuming from row {start_idx + 1}")
except FileNotFoundError:
    # If no output file exists, start fresh
    start_idx = 0
    print("Starting from the beginning.")

# Create a list for release dates to append
release_dates = df['release_date'].tolist() if start_idx > 0 else []

# Loop through DataFrame starting from the last processed row
for idx, row in df.iloc[start_idx:].iterrows():
    track_id = row['track_id']

    # Get the release date for the current track
    release_date = get_release_date(track_id, access_token)

    # Append the release date to the list
    release_dates.append(release_date)

    # Update the release date in the DataFrame
    df.at[idx, 'release_date'] = release_date

    # Print progress
    print(f"Processing track {idx + 1}/{len(df)}")

    # Sleep to avoid rate limit issues
    time.sleep(0.4)

    # Optionally save the DataFrame every 100 iterations
    if idx % 100 == 0:
        df.to_csv('output.csv', index=False)
        print(f"Progress saved after {idx + 1} tracks.")

# After the loop, save the final DataFrame
df.to_csv('output.csv', index=False)
print("Final DataFrame saved.")
```
